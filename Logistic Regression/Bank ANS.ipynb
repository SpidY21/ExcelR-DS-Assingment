```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\Yash\Desktop\CodeBase\EXCELR\Assingments\Answers\Logistic Regression\bank-full.csv")
```

```python
dataset.head()
```

```python
df_job = pd.get_dummies(dataset.job)
df_job
```

```python
df_job = df_job.replace(False,0).replace(True,1)
df_job
```

```python
df_j = df_job.replace(False,0).replace(True,1)
```

```python
df_marital = pd.get_dummies(dataset.marital)
df_marital = df_marital.replace(False,0).replace(True,1)
df_marital
```

```python
df_education = pd.get_dummies(dataset.education)
df_education = df_education.replace(False,0).replace(True,1)
df_education
```

```python
df_contact = pd.get_dummies(dataset.contact)
df_contact = df_contact.replace(False,0).replace(True,1)
df_contact
```

```python
df_outcome = pd.get_dummies(dataset.outcome)
df_outcome = df_outcome.replace(False,0).replace(True,1)
df_outcome
```

```python
dataset1 = pd.concat([df_contact,df_education,df_job,df_marital,df_outcome,dataset],axis=1)
```

```python
dataset1.head()
```

```python
dataset1.info()
```

```python
dataset1 = dataset1.drop('job',axis=1)
dataset1 = dataset1.drop('marital',axis=1)
dataset1 = dataset1.drop('education',axis=1)
dataset1 = dataset1.drop('default',axis=1)
dataset1 = dataset1.drop('contact',axis=1)
```

```python
dataset1.housing = dataset1.housing.replace('yes',1).replace('no',0)
```

```python
dataset1.info()
```

```python
dataset1.head()
```

```python
dataset1 = dataset1.drop('outcome',axis=1)
```

```python
dataset1 = dataset1.drop('month',axis=1)
dataset1 = dataset1.drop('day',axis=1)
dataset1.loan = dataset1.loan.replace('yes',1).replace('no',0)
```

```python
dataset1.head()
```

```python
dataset1.info()
```

```python
dataset1.y = dataset1.y.replace('no',0).replace('yes',1)
```

```python
X = dataset1.iloc[:, :-1].values
y = dataset1.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
print("The accuracy for the model is: ",accuracy_score(y_test,y_pred)*100)
```
