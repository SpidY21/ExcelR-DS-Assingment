```python
numeric_features = ['hair', 'feathers', 'eggs', 'milk', 'airborne', 'aquatic', 'predator', 'toothed', 'backbone', 'breathes', 'venomous', 'fins', 'legs', 'tail', 'domestic', 'catsize']
plt.figure(figsize=(16, 10))
for feature in numeric_features:
    plt.subplot(4, 4, numeric_features.index(feature) + 1)
    sns.boxplot(x='type', y=feature, data=dataset)
    plt.xlabel('Animal Type')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

```python
df = dataset.drop('animal name',axis=1)
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy = accuracy_score(y_test, y_pred) * 100
```

```python
print("The Accuracy of the model is: ",accuracy)
```
